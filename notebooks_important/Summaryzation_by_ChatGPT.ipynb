```python
def ChatGPT_Samary(OAI_API_KEY, company_name, message_text, sentiment_value):
    """
    Summarizes the provided message_text about the specified company_name with a given sentiment_value.

    Parameters:
        OAI_API_KEY (str): The OpenAI API key for authentication.
        company_name (str): The name of the company mentioned in the message_text.
        message_text (str): The text of the news article or message to be summarized.
        sentiment_value (int): The sentiment value associated with the message_text.

    Returns:
        str: A summary of the message_text considering:
            1. A brief overview of the part of the text discussing the specified company.
            2. The given sentiment_value towards the company in digital equivalent.
            3. An AI-generated rating (from 1 to 5) of the emotional tone of the message towards the company in digital equivalent.
    """
    prompt = f"""
            ВВОДНЫЕ ДАННЫЕ: 
            \nУ нас есть текст новости, о компании {company_name}, 
            \nон имеет эмоциональную оценку {sentiment_value}, 
            \nсам текст: \n{message_text}. 

            \n\n\n
            ЗАДАНИЕ:
            \nТебе нужно сделать саммари этого текста с учетом: 
            \n1. Вывести краткое содержание части текста которая говорит об этой компании
            \n2. Вывести заданную эмоциональную окрасу по отношению к этой компании в цифровом эквиваленте 
            \n3. Вывести свою оценку (от 1 до 5) эмоционального окраса этого сообщения по отношению к этой компании  цифровом эквиваленте
            """
    # print(prompt)

    client = OpenAI(api_key=OAI_API_KEY)

    completion = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
        {"role": "user", "content": prompt}
        ]
    )

    # print(completion.choices[0].message)
    return completion.choices[0].message
```
